```
ep = 8.8541878176e-12
l=5 

def f(x):                                       #def. función dependiente x
  return ((x)/(((x**2)+(y**2))))  
def g(y):                                       #def. función dependiente y
  return ((y)/(((x**2)+(y**2))))

N = 10
a=-5                                           #limites integrales
b=5
c=-5
d=5
hx=(b-a)/N
hy=(c-b)/N
```

```
px=0
py=0
ix=0
iy=0
Ix=0
Iy=0


                                               #Solucionar integral por Simpson
for k in range(1,N,2):                         #pares
  ix += f(a+(k*hx))                         
  iy += g(c+(k*hy))

for k in range (2,N,2):                        #impares
  px += f(a+(k*hx))
  py += g(c+(k*hy))

Ix=(hx/3)*(f(a)+f(b)+(4*ix)+(2*px))           #Solución integrales
Iy=(hy/3)*(g(c)+g(d)+(4*iy)+(2*py))

E=(l/(4*pi*ep))*(Ix+Iy)                       #Total campo eléctrico
X = linspace(-5,5,500)                        #arreglo que va de -5 hasta 5 con 500 pasos
Y = linspace(-5,5,500)
x,y=meshgrid(X,Y)                             #crea matriz de coordenadas apartir de los vectores


plt.contourf(y,x,E,cmap='jet')               #plot densidad de carga
plt.colorbar()
plt.show()
    
```
